```python
def ConstBoard(board):
    print("Current State of board :\n")
    for i in range(0,3):
        for j in range(0,3):
            if board[i][j] == 0:
                print("-",end=" ")
            if board[i][j] == 1:
                print('O',end=" ")  # Changed '0' to 'O' for clarity
            if board[i][j] == -1:
                print("X",end=" ")
            print("\n")

def UserTurn(board):
    r = int(input("Enter X's position in row: "))
    c = int(input("Enter X's position in column: "))

    while board[r][c] != 0:
        print("Wrong Move!!!")
        r = int(input("Enter X's position in row: "))
        c = int(input("Enter X's position in column: "))

    board[r][c] = -1

def check_winner(board):
    # checking row win
    for i in range(0, 3):
        if board[i][0] == board[i][1] == board[i][2] != 0:
            return board[i][0]
    # checking column win
    for i in range(0, 3):
        if board[0][i] == board[1][i] == board[2][i] != 0:
            return board[0][i]
    # checking diagonal win
    if board[0][0] == board[1][1] == board[2][2] != 0:
        return board[0][0]
    if board[0][2] == board[1][1] == board[2][0] != 0:
        return board[0][2]

    return 0

def minimax(board, player):
    x = check_winner(board)
    if x != 0:
        return x * player
    posr = -1
    posc = -1
    value = -2
    for i in range(0,3):
        for j in range(0,3):
            if board[i][j] == 0:
                board[i][j] = player
                score = minimax(board, (player * -1))
                if score > value:
                    value = score
                    posr = i
                    posc = j
                board[i][j] = 0
    if posr == -1 and posc == -1:
        return 0
    return value

def CompTurn(board):
    posr = -1
    posc = -1
    value = -2
    for i in range(0,3):
        for j in range(0,3):
            if board[i][j] == 0:
                board[i][j] = 1
                score = minimax(board, -1)
                board[i][j] = 0
                if score > value:
                    value = score
                    posr = i
                    posc = j
    board[posr][posc] = 1

player = int(input("Enter to play 1st(1) or 2nd(2) : "))
board = [[0,0,0],[0,0,0],[0,0,0]]

while check_winner(board) == 0:
    if player == 1:
        ConstBoard(board)
        UserTurn(board)
        if check_winner(board) != 0:
            break
        CompTurn(board)
    else:
        CompTurn(board)
        if check_winner(board) != 0:
            break
        ConstBoard(board)
        UserTurn(board)

x = check_winner(board)
ConstBoard(board)
if x == 0:
    print("Draw !!!")
elif x == -1:
    print("X Wins !!! Y loses !!")
elif x == 1:
    print("X loses!!! O wins")

    
```
